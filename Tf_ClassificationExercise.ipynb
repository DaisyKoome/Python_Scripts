```python
#Import
import pandas as pd
```

```python
#Reading in the census data
census=pd.read_csv('census_data.csv')
```

```python
census.head()
```

```python
#Converting the income bracket column from strings to 0s and 1s
census['income_bracket'].unique()
```

```python
#Creating a function that replaces the strings with boolean values
def label_fix(label):
    if label==' <=50K':
        return 0
    else:
        return 1
```

```python
#Applying the function to the column
census['income_bracket']=census['income_bracket'].apply(label_fix)
```

```python
census.head()
```

```python
#Assigning data for x and y axis
x_info=census.drop('income_bracket',axis=1)
label=census['income_bracket']
```

```python
#Performing a train-test split
#Imports
import tensorflow as tf
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x_info,label,test_size=0.3,random_state=101)
```

```python
#The columns in the data
census.columns
```

```python
#Creating feature columns
#Feature columns for categorical values using hash-buckets
w_class=tf.feature_column.categorical_column_with_hash_bucket('workclass',hash_bucket_size=5)
ed=tf.feature_column.categorical_column_with_hash_bucket('education',hash_bucket_size=15)
m_status=tf.feature_column.categorical_column_with_hash_bucket('marital_status',hash_bucket_size=5)
occ=tf.feature_column.categorical_column_with_hash_bucket('occupation',hash_bucket_size=20)
rel=tf.feature_column.categorical_column_with_hash_bucket('relationship',hash_bucket_size=10)
race=tf.feature_column.categorical_column_with_hash_bucket('race',hash_bucket_size=10)
gen=tf.feature_column.categorical_column_with_hash_bucket('gender',hash_bucket_size=2)
country=tf.feature_column.categorical_column_with_hash_bucket('native_country',hash_bucket_size=10)
```

```python
#Continuous feature columns for continuous values using numeric column
age=tf.feature_column.numeric_column('age')
ed_num=tf.feature_column.numeric_column('education_num')
capital_gain=tf.feature_column.numeric_column('capital_gain')
capital_loss=tf.feature_column.numeric_column('capital_loss')
hours_per_week=tf.feature_column.numeric_column('hours_per_week')
```

```python
#Compiling all feature columns into one list
f_cols=[w_class,ed,m_status,occ,rel,race,gen,country,age,ed_num,capital_gain,capital_loss,hours_per_week]
```

```python
#Input function
input_func=tf.estimator.inputs.pandas_input_fn(X_train,y_train,batch_size=100,num_epochs=1000,shuffle=True)
```

```python
#The model
model=tf.estimator.LinearClassifier(feature_columns=f_cols,n_classes=2)
```

```python
#Training the model
model.train(input_fn=input_func,steps=10000)
```

```python
#Evaluation by predicting
#Prediction input function
pred_input_fn=tf.estimator.inputs.pandas_input_fn(x=X_test,batch_size=100,num_epochs=1,shuffle=False)
```

```python
#Calling the predict function
predictions=model.predict(pred_input_fn)
```

```python
#Passing the predictions into a list
preds=list(predictions)
```

```python
preds
```

```python
#List of class ids from the predicted list to be compared against the real y values
final_preds=[]
for pred in preds:
    final_preds.append(pred['class_ids'][0])
```

```python
final_preds[:10]
```

```python
#Classification report
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,final_preds))
```
