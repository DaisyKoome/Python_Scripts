```python
import pandas as pd
```

```python
#Import the housing data and store in a variable
housing_data=pd.read_csv('cal_housing_clean.csv')
```

```python
#Viewing the data
housing_data.head()
```

```python
#Assigning data to the x and y axis
x_content=housing_data.drop('medianHouseValue',axis=1)
x_content.head()
y_value=housing_data['medianHouseValue']
```

```python
#Separating data into a training set and a testing set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x_content,y_value,test_size=0.3,random_state=101)
```

```python
#Getting a description of the data
housing_data.describe()
```

```python
#Using sklearn to create a MinMaxScaler for the feature data- scaling the feature data
from sklearn.preprocessing import MinMaxScaler
```

```python
#Storing it in a variable called scaler
scaler=MinMaxScaler()
```

```python
#Fitting it to the training data
scaler.fit(X_train)
```

```python
#Transforming X_train and X_test using the scaler and recreating them into data frames of scaled data using pd.Dataframe
X_train=pd.DataFrame(data=scaler.transform(X_train),columns=X_train.columns,index=X_train.index)
```

```python
X_test=pd.DataFrame(data=scaler.transform(X_test),columns=X_test.columns,index=X_test.index)
```
